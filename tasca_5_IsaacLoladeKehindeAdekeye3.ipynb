```python
import datashader as ds
import datashader.transfer_functions as tf
import pandas as pd
import numpy as np
import dask.dataframe as dd

datos = pd.read_csv("https://opendata-ajuntament.barcelona.cat/resources/bcn/Arbrat/OD_Arbrat_Viari_BCN.csv")
limpio = datos.dropna(subset=['x_etrs89', 'y_etrs89'])
cvs = ds.Canvas(plot_width=1000, plot_height=1000, x_range=(limpio['x_etrs89'].min(),limpio['x_etrs89'].max()), y_range=(limpio['y_etrs89'].min(),limpio['y_etrs89'].max()))
grafico = cvs.points(limpio, 'x_etrs89', 'y_etrs89')
tf.shade(grafico, cmap=['white', 'green', 'orange'])
```
